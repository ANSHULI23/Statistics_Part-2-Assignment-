## **1. What is hypothesis testing in statistics?**  
Hypothesis testing is a statistical method used to determine whether a certain claim about a population is supported by sample data. It involves comparing observed results with expected outcomes to make data-driven decisions.

### **Example:**  
A researcher wants to know if a new teaching method improves students' math scores. By conducting hypothesis testing, they compare test results from students using the new method versus those using the traditional approach.

---

## **2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**  
- **Null Hypothesis (H₀):** Assumes no effect or no difference; it represents the status quo.  
- **Alternative Hypothesis (H₁ or Ha):** Represents a new claim that contradicts H₀ and suggests a significant effect or difference.

### **Example:**  
A company claims that its new weight-loss supplement helps users lose more than 5 kg in a month.  
- **H₀:** The supplement does not help in losing more than 5 kg (μ ≤ 5 kg).  
- **H₁:** The supplement helps in losing more than 5 kg (μ > 5 kg).

---

## **3. What is the significance level in hypothesis testing, and why is it important?**  
The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true. It is usually set at 0.05 (5%), meaning that there is a 5% risk of making a Type 1 error.

### **Importance:**  
- Helps determine the threshold for statistical significance.  
- A lower α (e.g., 0.01) makes the test more strict but reduces the chance of a false positive.  
- A higher α (e.g., 0.10) makes the test more lenient, increasing the risk of a false positive.

---

## **4. What does a P-value represent in hypothesis testing?**  
The **P-value** represents the probability of obtaining results as extreme as the observed results, assuming that the null hypothesis is true.  

### **Interpretation:**  
- **Low P-value (≤ α):** Strong evidence against H₀ → Reject the null hypothesis.  
- **High P-value (> α):** Weak evidence against H₀ → Fail to reject the null hypothesis.

---

## **5. How do you interpret the P-value in hypothesis testing?**  
The P-value helps in decision-making:  
- **P ≤ 0.05:** Statistically significant → Reject H₀.  
- **P > 0.05:** Not statistically significant → Fail to reject H₀.

### **Example:**  
A study tests if a new fertilizer improves crop yield. If **P = 0.02** and α = 0.05, then **P < α**, so we reject H₀ and conclude the fertilizer improves yield.

---

## **6. What are Type 1 and Type 2 errors in hypothesis testing?**  
- **Type 1 Error (False Positive):** Rejecting H₀ when it is actually true.  
- **Type 2 Error (False Negative):** Failing to reject H₀ when it is actually false.

### **Example:**  
- **Type 1 Error:** A doctor diagnoses a healthy patient with a disease.  
- **Type 2 Error:** A doctor fails to diagnose an ill patient.

---

## **7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**  
- **One-tailed test:** Tests for an effect in only one direction (e.g., "greater than" or "less than").  
- **Two-tailed test:** Tests for an effect in both directions (e.g., "different from" without specifying more or less).

### **Example:**  
- **One-tailed:** Checking if a new medicine increases blood pressure.  
- **Two-tailed:** Checking if a new medicine changes blood pressure in any direction.

---

## **8. What is the Z-test, and when is it used in hypothesis testing?**  
The **Z-test** is used when:  
- The sample size is large (n > 30).  
- The population variance (σ²) is known.  

It helps compare sample means to population means.

### **Example:**  
A school believes the average IQ of its students is 100. A researcher takes a sample of 50 students and conducts a **Z-test** to check if the sample's mean IQ significantly differs from 100.

---

## **9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**  
The **Z-score** tells how many standard deviations a sample mean is from the population mean.  

### **Formula:**  
$$ Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}} $$

Where:  
- **X̄** = Sample mean  
- **μ** = Population mean  
- **σ** = Population standard deviation  
- **n** = Sample size  

### **Example:**  
A company claims its employees work **40 hours per week**. A random sample of 50 employees shows an average of **42 hours** with a standard deviation of **5 hours**.  

$$ Z = \frac{{42} - \ 40}{\frac{5}{\sqrt{50}}} = 2.83 $$

Since **Z = 2.83** is significant at α = 0.05, we conclude that employees work significantly more than 40 hours.

---

## **10. What is the T-distribution, and when should it be used instead of the normal distribution?**  
The **T-distribution** is used when:  
- The sample size is small (n < 30).  
- The population standard deviation (σ) is unknown.  

It has heavier tails than the normal distribution, meaning it accounts for variability in small samples.

### **Example:**  
A scientist wants to compare the test scores of **15 students** before and after using a new study technique. Since the sample size is small and σ is unknown, a **T-distribution** is used.

| Feature | **Z-Test** | **T-Test** |

| **Distribution Used** | Normal | T-distribution |

## **12. What is the T-test, and how is it used in hypothesis testing?**  
A **T-test** compares the means of two groups when the population variance is unknown.

### **Types of T-tests:**  
1. **One-sample T-test:** Compares a sample mean to a known population mean.  
2. **Independent T-test:** Compares two different sample groups.  
3. **Paired T-test:** Compares two measurements from the same group (before-after studies).

### **Example:**  
A researcher tests if students using a new study method score higher than those using a traditional method. A **T-test** is conducted to compare their mean scores.

---

## **13. What is the relationship between Z-test and T-test in hypothesis testing?**  
- The **T-test** is a generalization of the **Z-test** for smaller samples.  
- As sample size increases, the T-distribution approximates the normal distribution, making the Z-test more applicable.

### **Example:**  
A researcher studies the **average height** of people in two cities.  
- If **n > 30**, a **Z-test** is used.  
- If **n < 30**, a **T-test** is preferred.

## **14. What is a confidence interval, and how is it used to interpret statistical results?**  
A **confidence interval (CI)** provides a range in which a population parameter likely falls.

### **Formula:**  
$$
CI =\bar{X} \pm (Z \times \frac{\sigma}{\sqrt{n}})
$$

### **Example:**  
A survey finds that **average customer satisfaction** is **75%** with a **95% CI of (72%, 78%)**. This means we are **95% confident** that the true satisfaction rate is between **72% and 78%**.

---

## **15. What is the margin of error, and how does it affect the confidence interval?**  
The **margin of error (ME)** is the maximum expected difference between the sample estimate and the true population value. It determines the width of the confidence interval.  

### **Formula:**  
$$
ME = Z \times \frac{\sigma}{\sqrt{n}}
$$
Where:  
- **Z** = Critical value based on confidence level  
- **σ** = Population standard deviation  
- **n** = Sample size  

### **Effect on Confidence Interval:**  
- A **larger margin of error** results in a **wider confidence interval**, making the estimate less precise.  
- A **smaller margin of error** results in a **narrower confidence interval**, making the estimate more precise.  

### **Example:**  
If a poll shows **60%** approval for a policy with a **margin of error of ±3%**, the confidence interval is **(57%, 63%)**, meaning the true approval rate is likely within this range.

---

## **16. How is Bayes' Theorem used in statistics, and what is its significance?**  
**Bayes' Theorem** helps update the probability of an event based on new evidence. It is widely used in decision-making, medical testing, and machine learning.  

### **Formula:**  
$$
P(A|B) = \frac{P(B|A) P(A)}{P(B)}
$$
Where:  
- **P(A|B)** = Probability of event A given B has occurred  
- **P(B|A)** = Probability of event B given A has occurred  
- **P(A)** = Prior probability of A  
- **P(B)** = Prior probability of B  

### **Example:**  
In medical testing, if a patient tests positive for a rare disease, Bayes' Theorem helps determine the actual probability of having the disease based on the test's accuracy and the disease's prevalence.

---

## **17. What is the Chi-square distribution, and when is it used?**  
The **Chi-square (χ²) distribution** is used for categorical data analysis, particularly in **goodness of fit tests** and **independence tests**.

### **Applications:**  
- **Testing independence**: Checking if two categorical variables are related (e.g., smoking habits vs. lung disease).  
- **Goodness of fit**: Comparing observed data to an expected distribution.

### **Example:**  
A store wants to check if customer purchases are evenly distributed across the week. A **Chi-square test** compares observed sales data with expected sales distribution.

---

## **18. What is the Chi-square goodness of fit test, and how is it applied?**  
The **Chi-square goodness of fit test** determines whether sample data matches an expected distribution.

### **Formula:**  
$$ [\chi^2 = \sum \frac{(O - E)^2}{E}]  $$
Where:  
- **O** = Observed frequency  
- **E** = Expected frequency  

### **Example:**  
A company wants to know if its market share distribution matches its expectation. If **χ² is large**, it suggests that the actual market share differs significantly from the expected distribution.

---

## **19. What is the F-distribution, and when is it used in hypothesis testing?**  
The **F-distribution** is used to compare variances between two or more groups, often in **ANOVA** and **regression analysis**.

### **Key Characteristics:**  
- Positively skewed  
- Used in tests that involve variance comparison  

### **Example:**  
A researcher compares **three different fertilizers' effectiveness** on crop yield. The **F-test** helps determine if the variance in crop yield among fertilizers is statistically significant.

---

## **20. What is an ANOVA test, and what are its assumptions?**  
**ANOVA (Analysis of Variance)** is used to compare means of **three or more groups** to see if at least one is significantly different.

### **Assumptions:**  
1. **Independence**: Observations must be independent.  
2. **Normality**: Data should follow a normal distribution.  
3. **Homogeneity of variance**: Variances across groups should be similar.

### **Example:**  
A school tests if three different teaching methods impact student performance. **ANOVA** compares students' scores across the three groups.

---

## **21. What are the different types of ANOVA tests?**  
1. **One-way ANOVA**: Compares means of **one factor** (e.g., three diets affecting weight loss).  
2. **Two-way ANOVA**: Compares means of **two factors** (e.g., diets and exercise affecting weight loss).  
3. **Repeated measures ANOVA**: Compares means of the **same subjects over time** (e.g., student performance before and after training).

### **Example:**  
A company tests if **three different marketing strategies** result in different sales. **One-way ANOVA** is used to check for significant differences.

---

## **22. What is the F-test, and how does it relate to hypothesis testing?**  
The **F-test** compares the variances of two or more groups to check if they are significantly different.

### **Formula:**  
$$ [F = \frac{\text{Variance of Group 1}}{\text{Variance of Group 2}}] $$  

### **Relation to Hypothesis Testing:**  
- Used in **ANOVA** to compare multiple group means.  
- Used in **regression analysis** to test model significance.  

### **Example:**  
A factory compares the **variability in defect rates** across different machine operators. If the F-test shows a significant difference, it suggests some operators have more consistent performance than others.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

# Given data
sample = np.array([52, 48, 55, 53, 49, 51, 54, 50, 47, 56])
pop_mean = 50
pop_std = 3
n = len(sample)

# Compute Z-score
sample_mean = np.mean(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# Compute P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (Significant difference).")
else:
    print("Fail to reject the null hypothesis (No significant difference).")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample data
np.random.seed(42)
sample = np.random.normal(50, 10, 30)
population_mean = 50
population_std = 10
n = len(sample)

# Compute Z-score
z_score = (np.mean(sample) - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Visualization
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y, label="Standard Normal Distribution")
plt.axvline(z_score, color="r", linestyle="--", label=f"Z = {z_score:.2f}")
plt.fill_between(x, y, where=(x > abs(z_score)) | (x < -abs(z_score)), color='red', alpha=0.3)
plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.show()

print(f"Z-score: {z_score}, P-value: {p_value}")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    null_dist = norm.pdf(x, 0, 1)
    alt_dist = norm.pdf(x, 2, 1)

    plt.plot(x, null_dist, label="Null Distribution (H0)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Distribution (H1)", color="green")

    critical_value = norm.ppf(1 - alpha)
    plt.axvline(critical_value, color="red", linestyle="--", label="Critical Value (α)")
    plt.fill_between(x, null_dist, where=(x > critical_value), color="red", alpha=0.3, label="Type I Error (α)")
    plt.fill_between(x, alt_dist, where=(x < critical_value), color="purple", alpha=0.3, label="Type II Error (β)")

    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_errors()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = stats.sem(sample)
    margin_error = std_err * stats.t.ppf((1 + confidence) / 2, len(sample) - 1)
    return mean - margin_error, mean + margin_error

print(confidence_interval(np.random.normal(50, 10, 30)))
```

```python
#11. Implement o Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Given probabilities
prior_prob = 0.01  # Prior probability (e.g., probability of a disease)
sensitivity = 0.95  # Probability of a positive test given disease
false_positive_rate = 0.05  # Probability of a positive test without disease

# Compute posterior probability using Bayes' Theorem
numerator = sensitivity * prior_prob
denominator = (sensitivity * prior_prob) + (false_positive_rate * (1 - prior_prob))
posterior_prob = numerator / denominator

# Display Results
print(f"Updated Probability after Testing (Posterior Probability): {posterior_prob}")
```

```python
#16. Implement on F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats

# Generate random data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(50, 15, 30)

# Compute variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# F-test statistic
f_stat = var1 / var2
df1, df2 = len(group1)-1, len(group2)-1
p_value = stats.f.cdf(f_stat, df1, df2)

# Display Results
print(f"F-Statistic: {f_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.
import numpy as np
import scipy.stats as stats

# Generate sample data for 3 groups
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Display Results
print(f"F-Statistic: {f_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in means.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

from scipy.stats import shapiro, levene

# Check Normality
print("Normality Test (Shapiro-Wilk):")
for i, group in enumerate([group1, group2, group3], start=1):
    stat, p = shapiro(group)
    print(f"Group {i}: p-value = {p}")

# Check Homogeneity of Variance
stat, p = levene(group1, group2, group3)
print(f"\nLevene’s Test for Variance: p-value = {p}")

if p < 0.05:
    print("Reject null hypothesis: Variances are not equal.")
else:
    print("Fail to reject null hypothesis: Variances are equal.")
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Generate random sample data
np.random.seed(42)
group_A_X = np.random.normal(50, 10, 15)
group_B_X = np.random.normal(55, 10, 15)
group_A_Y = np.random.normal(52, 10, 15)
group_B_Y = np.random.normal(58, 10, 15)

# Create DataFrame
df = pd.DataFrame({
    'Value': np.concatenate([group_A_X, group_B_X, group_A_Y, group_B_Y]),
    'Factor1': ['A']*15 + ['B']*15 + ['A']*15 + ['B']*15,
    'Factor2': ['X']*30 + ['Y']*30
})

# Perform Two-Way ANOVA
model = ols('Value ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display Results
print(anova_table)
```

```python
#22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Generate random data
np.random.seed(42)
group_A = np.random.normal(50, 10, 15)
group_B = np.random.normal(55, 10, 15)
group_C = np.random.normal(60, 10, 15)

# Create DataFrame
df = pd.DataFrame({
    'Value': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A']*15 + ['B']*15 + ['C']*15
})

# Perform One-Way ANOVA
model = ols('Value ~ C(Group)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA table
print(anova_table)

# Boxplot Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Value', data=df)
plt.title('One-Way ANOVA Boxplot')
plt.show()
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

from scipy.stats import ttest_ind

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(55, 10, 30)
sample2 = np.random.normal(60, 10, 30)

# Perform an independent T-test
t_stat, p_value = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Conclusion
if p_value < 0.05:
    print("Reject null hypothesis: Means are significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference.")
```

```python
#24. Perform a hypothesis test for population variarice using a Chi-square distribution and interpret the results.

from scipy.stats import chi2

# Given variance and hypothesized variance
np.random.seed(42)
sample = np.random.normal(50, 10, 30)
sample_variance = np.var(sample, ddof=1)
n = len(sample)
hypothesized_variance = 100  # Assume population variance

# Chi-square test statistic
chi_square_stat = (n - 1) * sample_variance / hypothesized_variance

# Critical values (5% significance level)
alpha = 0.05
chi_critical_low = chi2.ppf(alpha / 2, df=n-1)
chi_critical_high = chi2.ppf(1 - alpha / 2, df=n-1)

print(f"Chi-square Statistic: {chi_square_stat:.3f}")
print(f"Critical Range: ({chi_critical_low:.3f}, {chi_critical_high:.3f})")

if chi_square_stat < chi_critical_low or chi_square_stat > chi_critical_high:
    print("Reject null hypothesis: Variance is significantly different.")
else:
    print("Fail to reject null hypothesis: Variance is not significantly different.")
```

```python
#25. Write a Python script to perform a 2-test for comparing proportions between two datasets or groups.

from statsmodels.stats.proportion import proportions_ztest

# Sample data (successes, total observations)
successes = np.array([120, 150])
nobs = np.array([300, 350])

# Perform Z-test for proportions
z_stat, p_value = proportions_ztest(successes, nobs)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject null hypothesis: Proportions are significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference in proportions.")
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
from scipy.stats import f

# Generate random data
np.random.seed(42)
sample1 = np.random.normal(50, 10, 30)
sample2 = np.random.normal(55, 15, 30)

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-test statistic
f_stat = var1 / var2
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Critical F-value
alpha = 0.05
f_critical_low = f.ppf(alpha / 2, df1, df2)
f_critical_high = f.ppf(1 - alpha / 2, df1, df2)

print(f"F-Statistic: {f_stat:.3f}")
print(f"Critical Range: ({f_critical_low:.3f}, {f_critical_high:.3f})")

if f_stat < f_critical_low or f_stat > f_critical_high:
    print("Reject null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference in variances.")
```
